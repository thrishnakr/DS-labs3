```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords') #preposition words
nltk.download('punkt')# punctuations
nltk.download('averaged_perceptron_tagger')#to download tags

stop_words = set(stopwords.words('english'))
  
#Dummy text
txt = "Hello. MCA S3 is fantastic. We learn many new concepts and implement them in our practical exams. "\
"1st of all the data science is a new paper."
  
# sent_tokenize is one of instances of 
# PunktSentenceTokenizer from the nltk.tokenize.punkt module
  
tokenized = sent_tokenize(txt) # parting the sentences
for i in tokenized: # i is words
      
    # Word tokenizers is used to find the words 
    # and punctuation in a string
    wordsList = nltk.word_tokenize(i)
  
    # removing stop words from wordList
    wordsList = [w for w in wordsList if not w in stop_words] 
  
    #  Using a Tagger. Which is part-of-speech 
    # tagger or POS-tagger. 
    tagged = nltk.pos_tag(wordsList)
  
    print(tagged)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
#1st letter is not considered as stop word(whether it is 'is','the','an' etc..)
```

```python
#method to generate n-grams:
#params:
#text-the text for which we have to generate n-grams
#ngram-number of grams to be generated from the text(1,2,3,4 etc., default value=1)
def generate_N_grams(text,ngram=1):
  words=[word for word in text.split(" ") if word not in set(stopwords.words('english'))]  
  print("Sentence after removing stopwords:",words)
  temp=zip(*[words[i:] for i in range(0,ngram)]) # based on size of ngram, words are combined together
  ans=[' '.join(ngram) for ngram in temp]
  return ans
```

```python
generate_N_grams("The sun rises in the east",2)
```

```python
generate_N_grams("The sun rises in the east",3)
```

```python
generate_N_grams("The sun rises in the east",4)
```
